```python
def change_pm(pm2, pm10):
    if pm2 < 35 :
        a = '정상'
    elif (pm2 >= 35 and pm2 < 75) :
        a = "나쁨"
    else :
        a = "매우 나쁨"
        
    if pm10 < 80 :
        b = '정상'
    elif (pm10 >= 80 and pm10 < 150) :
        b = "나쁨"
    else :
        b = "매우 나쁨"
        
    if (a == "매우 나쁨") or (b == "매우 나쁨") :
        result = "매우 나쁨"
    elif (a == "나쁨") or (b == "나쁨") :
        result = "나쁨"
    else :
        result = '정상'
        
    return result
```

```python
def change_co2(x):
    if x < 1000 :
        result = '정상'
    elif (x >= 1000 and x < 1500) :
        result = "나쁨"
    else :
        result = "매우 나쁨"
        
    return result
```

```python
def change_tvoc(x):
    if x < 660 :
        result = '정상'
    elif (x >= 660 and x < 2200) :
        result = "나쁨"
    else :
        result = "매우 나쁨"
        
    return result
```

```python
# 미세먼지(실내)
df['c_pm'] = df.apply(lambda x : change_pm(x['pm2'], x['pm10']), axis=1)

# 미세먼지(실외)
df['c_O_pm'] = df.apply(lambda x : change_pm(x['O_pm2'], x['O_pm10']), axis=1)

# CO2(실내)
df['c_CO2'] = df['CO2'].apply(lambda x : change_co2(x))

# TVOC실내)
df['c_tvoc'] = df['tvoc'].apply(lambda x : change_tvoc(x))
```
